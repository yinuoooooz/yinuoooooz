```python
import fitz  # PyMuPDF
import concurrent.futures

def extract_text_from_pdf(file_path):
    print(f"Processing {file_path}")
    try:
        text = ""
        with fitz.open(file_path) as doc:
            for page in doc:
                text += page.get_text()
        print(f"Finished processing {file_path}")
        return text
    except Exception as e:
        print(f"Failed to process {file_path}: {str(e)}")
        return ""

pdf_file_paths = [
    "/content/drive/MyDrive/files for ai/cma5_auv_10g_LnDfunding.pdf",
    "/content/drive/MyDrive/files for ai/cma5_auv_4_gst.pdf",
    "/content/drive/MyDrive/files for ai/cma5_auv_5_JTWP.pdf",
    "/content/drive/MyDrive/files for ai/cma5_auv_6_MWP.pdf",
    "/content/drive/MyDrive/files for ai/cma5_auv_8a_gga.pdf",
    "/content/drive/MyDrive/files for ai/cop28_auv_2h_youth.pdf",
    "/content/drive/MyDrive/files for ai/cop28_auv_8g_lnd.pdf",
    "/content/drive/MyDrive/files for ai/cop28_cma5_auv_2c_youth.pdf",
    # Make sure each path is a separate string in the list
]

# Collect all text in a single string
all_text = ""

# Use ThreadPoolExecutor to extract text concurrently
with concurrent.futures.ThreadPoolExecutor(max_workers=len(pdf_file_paths)) as executor:
    future_to_file = {executor.submit(extract_text_from_pdf, file_path): file_path for file_path in pdf_file_paths}
    for future in concurrent.futures.as_completed(future_to_file):
        try:
            text = future.result()
            if text:
                all_text += text + "\n\n"  # Separate text from different PDFs with new lines
        except Exception as e:
            print(f"Error processing PDF: {str(e)}")

# Save the collected text into a single file
output_file_path = "/content/drive/MyDrive/extracted_text.txt"
with open(output_file_path, "w") as text_file:
    text_file.write(all_text)

print(f"All text has been extracted and saved to {output_file_path}.")
```
